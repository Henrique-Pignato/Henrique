### Imoportação e Tratamento dos Dados

```python
import pandas as pd

cadastro_func = pd.read_csv(r'C:\Users\henri\OneDrive\Documentos\Portifólio\Pandas\CadastroFuncionarios.csv', sep=';', decimal=',')
cadastro_clientes = pd.read_csv(r'C:\Users\henri\OneDrive\Documentos\Portifólio\Pandas\CadastroClientes.csv', sep=';', decimal=',')
base_serv = pd.read_excel(r'C:\Users\henri\OneDrive\Documentos\Portifólio\Pandas\BaseServiçosPrestados.xlsx')

display(cadastro_func)
display(cadastro_clientes)
display(base_serv)
```

### 1) Valor Total Folha Salarial

```python
#Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa?
display(cadastro_func)
#cadastro_func.info()

cadastro_func['Gasto Funcionario'] = cadastro_func['Salario Base'] + cadastro_func['Impostos'] + cadastro_func['Beneficios'] + cadastro_func['VT'] + cadastro_func['VR']

Total_folha = 'O valor total da folha salarial é de R$ {:,}'.format(cadastro_func['Gasto Funcionario'].sum())


Total_folha = Total_folha.replace(',', '/')
Total_folha = Total_folha.replace('.', ',')
Total_folha = Total_folha.replace('/', '.')
print(Total_folha)
```

### 2) Faturamento Total

```python
#Qual foi o faturamento da empresa?
display(base_serv)
display(cadastro_clientes)

base_merge = pd.merge(
    base_serv, # <- primeira base
    cadastro_clientes, # <- segunda base
    how='inner', # <- tipo de junção que vamos fazer
    on="ID Cliente" # <- coluna que vamos usar para fazer essa junção das bases
)

#display(base_merge)

base_merge['Faturamento'] = base_merge['Tempo Total de Contrato (Meses)'] * base_merge['Valor Contrato Mensal']

Faturamento = 'O faturamento total é: R${:,}'.format(base_merge['Faturamento'].sum())
Faturamento = Faturamento.replace(',', '.')

print(Faturamento)
```

### 3) % de Funcionários

```python
#Qual o % de funcionários que já fechou algum contrato?

display(base_serv)
Funcionarios_unicos = base_serv['ID Funcionário'].unique().size
Funcionários_totais = cadastro_func['ID Funcionário'].size

print('A porcentagem de funcionários quevenderam foi de : {:.2%}'.format(Funcionarios_unicos / Funcionários_totais))
```

### 4) Total de Contratos de Cada Área

```python
#Calcule o total de contratos que cada área da empresa já fechou
#display(cadastro_func)
#display(base_serv)

merge_id = pd.merge(
    base_serv, # <- primeira base
    cadastro_func[['ID Funcionário' , 'Area']], # <- segunda base
    how='inner', # <- tipo de junção que vamos fazer
    on="ID Funcionário" # <- coluna que vamos usar para fazer essa junção das bases
)
#display(merge_id)
display(merge_id['Area'].value_counts())
```

### 5) Total de Funcionários por Área

```python
#Calcule o total de funcionários por área
display(cadastro_func)
cadastro_func['Area'].value_counts()
```

### 6) Ticket Médio

```python
#Qual o ticket médio mensal (faturamento médio mensal) dos contratos?

display(base_merge)
Ticket_medio = base_merge['Faturamento'].mean()

print(Ticket_medio)
```
